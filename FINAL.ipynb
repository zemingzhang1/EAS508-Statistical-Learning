#### Data Cleaning- we use Z-scores to remove true outliers:
Using Z-scores are helpful for identifying true outliers because they provide a standardized approach for measuring and comparing the degree of dispersion in a dataset. By using z-scores, you can easily identify data points that are too far away from the mean and are therefore likely to be true outliers. This is especially useful in datasets that have multiple dimensions, as z-scores make it easy to compare data points across each dimension. Furthermore, z-scores provide a quantitative method for detecting outliers and make it easier to decide which data points should be removed from a dataset.

````python
#Calculate z-score
threshold = 3
z_score = (df - df.mean()) / df.std()

#Identify outliers
outliers = df[(abs(z_score) > threshold).any(axis=1)]

#Remove outliers
df = df[~(abs(z_score) > threshold).any(axis=1)]
````

#### For Feature Selection:
The code below is a feature selection code using various methods for 3 different dataframes each containing a property. It first uses the correlation matrix to select highly correlated features with the output variable. Then, Chi-square tests are used to calculate the significance of the selected features and the p-values are used to determine which features are more relevant. After that, Ridge and Lasso regression are used to calculate the optimal alpha and the features associated with the alpha values are selected. Finally, the selected features are printed in order to compare the different methods.

````python
#Importing the necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import chi2
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV

idx = 0
targetsFS = {}

for k,v in targets.items():
    
    print('\n')
    print(f"Target Name: {k}")
    X, y = v
    #Calculate the correlation between the features
    corr_matrix =df.corr()
    #Correlation with output variable
    cor_target = abs(corr_matrix[k])
    #Selecting highly correlated features
    sorted(abs(num) for num in cor_target)[:5]
    relevant_features = cor_target[abs(cor_target)>0.2]
    try:
        corrFS = list(relevant_features.index)[1:].remove(k)[:5]
    except:
        corrFS = list(relevant_features.index)[1:][:5]

    print(f"Feature Selected by Corr: {corrFS}")
    
    #Chi-square test
    #apply chi2 test
    try:
        chi_score = chi2(X,y)
        #print the features selected and their performance
        p_values = pd.Series(chi_score[1], index=X.columns)
        p_values.sort_values(ascending = False , inplace = True)
        chiFS = p_values[p_values<0.2].index.values[:5]
        print('Features Selected by Chi:',chiFS)
        p_values.plot.bar()
        plt.show()
    except:
        chiFS = list(X.columns)[:5]
        print("Features Selected by Chi: The chi-squared test takes categorical label data as input.")
        
    
    #apply ridge
    regR = RidgeCV(cv=5).fit(X, y)
    #print the optimal alpha for ridge
    print('Optimal alpha for RIDGE :',regR.alpha_)
    
    #RIDGE
    #initialize Ridge Regression
    ridge = Ridge(alpha=regR.alpha_)
    #fit data
    ridge.fit(X,y)
    
    #print the features selected and their performance
    p_valuesR = pd.Series(ridge.coef_, index=X.columns)
    p_valuesR.sort_values(ascending = False , inplace = True)
    p_valuesR.plot.bar()
    plt.show()
    ridgeFS = p_valuesR[p_valuesR!=0].index.values
    print('Features Selected by RIDGE:',ridgeFS)
        
    #extract coefficients
    coefR = ridge.coef_
    feature_namesR = list(df.columns)
    feature_namesR.pop(idx)
    
    #select important features
    ridgeFS = [feature_namesR[i] for i in range(len(coefR)) if abs(coefR[i]) > 0.001][:5]
    print('Important Features by RIDGE:',ridgeFS)
    
    #apply LASSO
    regL = LassoCV(cv=5, random_state=0).fit(X, y)
    #print the optimal alpha for LASSO
    print('Optimal alpha for LASSO :',regL.alpha_)

    #LASSO
    reg = Lasso(alpha=regL.alpha_)
    #apply LASSO
    reg.fit(X,y)

    #print the features selected and their performance
    p_values = pd.Series(reg.coef_, index=X.columns)
    p_values.sort_values(ascending = False , inplace = True)
    p_values.plot.bar()
    plt.show()
    lassFS = p_values[p_values!=0].index.values
    print('Features Selected by LASSO:',lassFS)
    
    #extract coefficients
    coefL = reg.coef_
    feature_namesL = list(df.columns)
    feature_namesL.pop(idx)
    
    #select important features
    lassFS = [feature_namesL[i] for i in range(len(coefL)) if abs(coefL[i]) > 0.001][:5]
    print('Important Features by LASSO:',lassFS)
    
    
    targetsFS[k] = [(chiFS, "CHI"), (corrFS,"CORR"), (lassFS,"LASS"), (ridgeFS,"RIDG")]
    
    idx+=1
    print('\n')
````

#### For Feature Validation:
We now will validate the feature selection that was done previously for a selection of 3 different data frames. We split the data into a training and test set, and applies standard scaler to both the training and the test sets. Then, the Linear Regression model is fitted to the training data, the model is then used to predict the values for the test set, and the root mean squared error is calculated for each of the feature sets to evaluate the accuracy of the model. The root mean squared error (RMSE) measures how closely the predicted values from a model match the actual values. It is calculated as the square root of the mean of the squared differences between the predicted values and the actual values. In this case, the RMSE value is used to evaluate the accuracy of the Linear Regression model trained on the training set.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

summary = []
models = {}

for k, v in  targets.items():
    
    print(f"Target Name: {k}")
    fittingSet = []
    
    for fs,name in targetsFS[k]:
        print(f"After Feature Selection by {name}: {fs}")
        fs = list(fs)
        try:
            fs.remove(k)
        except:
            print("Property itself is filtered out.")
        X = df.filter(items=fs)
        display(X)
        # split df into trianing and testing set
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)

        # Feature Scaling
        sc_X = StandardScaler()
        X_train = sc_X.fit_transform(X_train)
        X_test = sc_X.transform(X_test)

        # Fitting  Linear Regression to the Training set
        regressor = LinearRegression()
        regressor.fit(X_train, y_train)

        # predicting the Test set results
        y_pred = regressor.predict(X_test)
        fittingSet.append((name,[X_train, X_test, y_train, y_test]))

        # calculate Root Mean Squared Error 
        try: 
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            print(f"Root Mean Squared Error of {name} of {k}: {rmse}")
            summary.append(f"Root Mean Squared Error of {name} of {k}: {rmse}")
        except: 
            print(f"Error calculating RMSE of {name} of {k}")
            
    models[k] = fittingSet 
````
